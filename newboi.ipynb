```python
from datasets import load_dataset
dataset = load_dataset("wikipedia", '20200501.en')
```

```python
import torch

if torch.cuda.is_available():
  dev = torch.device("cuda")
  print(f'There are {torch.cuda.device_count()} GPU(s) available.')
  print('Device name:', torch.cuda.get_device_name(0))
else:
  print('No GPU available, using the CPU instead.')
  dev = torch.device("cpu")
```

```python
import os
from tokenizers.models import BPE
from tokenizers import Tokenizer
from tokenizers.decoders import ByteLevel as ByteLevelDecoder
from tokenizers.normalizers import NFKC, Sequence
from tokenizers.pre_tokenizers import ByteLevel
from tokenizers.trainers import BpeTrainer

class BPE_token(object):
    def __init__(self):
        self.tokenizer = Tokenizer(BPE())
        self.tokenizer.normalizer = Sequence([
            NFKC()
        ])
        self.tokenizer.pre_tokenizer = ByteLevel()
        self.tokenizer.decoder = ByteLevelDecoder()

    def bpe_train(self, paths):
        #' &.@acbedgfihkjmlonqpsrutwvyxz
        trainer = BpeTrainer(vocab_size=31, show_progress=True, inital_alphabet=ByteLevel.alphabet(), special_tokens=[
            "<s>",
            "<pad>",
            "</s>",
            "<unk>",
            "<mask>"
        ])
        self.tokenizer.train(paths, trainer)

    def save_tokenizer(self, location, prefix=None):
        if not os.path.exists(location):
            os.makedirs(location)
        self.tokenizer.model.save(location, prefix)
```

```python
%cd /content/MiCT-RANet-ASL-FingerSpelling/data/
!mkdir text
%cd text/
!echo "abcdefghijklmnopqrstuvwxyz &@'." >> chars.txt
```

```python
from pathlib import Path
import os
# the folder 'text' contains all the files
# MAKE SURE TO CREATE THE FOLDER PRIOR TO RUNNING THIS AND UPLOAD THE CHARACTER TEXT FILE
paths = [str(x) for x in Path("./data/text/").glob("**/*.txt")]
tokenizer = BPE_token()
# train the tokenizer model
tokenizer.bpe_train(paths)
# saving the tokenized data in our specified folder 
save_path = 'tokenized_data'
tokenizer.save_tokenizer(save_path)
```

```python
import torch
from transformers import GPT2Config, GPT2LMHeadModel, GPT2Tokenizer
# loading tokenizer from the saved model path
tokenizer = GPT2Tokenizer.from_pretrained(save_path)
tokenizer.add_special_tokens({
  "eos_token": "</s>",
  "bos_token": "<s>",
  "unk_token": "<unk>",
  "pad_token": "<pad>",
  "mask_token": "<mask>"
})
# creating the configurations from which the model can be made
config = GPT2Config(
  vocab_size=tokenizer.vocab_size,
  gradient_checkpointing=True,
  use_cache=False,
  bos_token_id=tokenizer.bos_token_id,
  eos_token_id=tokenizer.eos_token_id
)
# creating the model
model = GPT2LMHeadModel(config)
```

```python
%cd /content/MiCT-RANet-ASL-FingerSpelling/data/
from transformers import WEIGHTS_NAME, CONFIG_NAME
import os
#print(os.getcwd())
#print ("pp")
output_dir = 'tokenized_data'
# creating directory if it is not present
if not os.path.exists(output_dir):
  print ("doesnt exist")
  os.mkdir(output_dir)
model_to_save = model.module if hasattr(model, 'module') else model
output_model_file = os.path.join(output_dir, WEIGHTS_NAME)
output_config_file = os.path.join(output_dir, CONFIG_NAME)
# save model and model configs
model.save_pretrained(output_dir)
model_to_save.config.to_json_file(output_config_file)
```

```python
%cd /content/MiCT-RANet-ASL-FingerSpelling/
tokenizer_save_path = "data/tokenized_data"
tokenizer = GPT2Tokenizer.from_pretrained(tokenizer_save_path)
```

```python
dataset = dataset['train']
```

```python
title_dataset = dataset['title']
```

```python
single_string = ''
for i in range(len(title_dataset)):
  single_string += title_dataset[i].lower() + tokenizer.eos_token
string_tokenized = tokenizer.encode(single_string)
```

```python
from torch.utils.data import TensorDataset, DataLoader

examples = []
block_size = 100
for i in range(0, len(string_tokenized) - block_size + 1, block_size):
  examples.append(string_tokenized[i:i + block_size])
inputs, labels = [], []
for ex in examples:
  inputs.append(torch.LongTensor(ex[:-1]))
  labels.append(torch.LongTensor(ex[1:]))
inputs = (torch.stack(inputs)).to(dev)
labels = (torch.stack(labels)).to(dev)
dataset = DataLoader(TensorDataset(inputs, labels), batch_size=600, shuffle=True)
```

```python
import torch.optim as optim
import torch.nn as nn

model = model.to(dev)

optimizer = optim.Adam(model.parameters(), lr=6e-4)
criterion = nn.CrossEntropyLoss()
```

```python
import torch.nn.functional as F
from torch.nn.utils import clip_grad_value_
import time

TIME_IN_SECONDS = 36000 # MAKE SURE TO CHANGE THIS TO THE TRUE VALUE YOU WANT

start = time.time()
done = False
accumulation_steps = 10 # To match GPT-3 Small's 500k bsz
model.zero_grad()

while not done:
    for i, (inputs, labels) in enumerate(dataset):
        outputs = model(inputs, labels=labels)
        loss = outputs.loss / accumulation_steps
        loss.backward()
        clip_grad_value_(model.parameters(), 1.0)
        if (i+1) % accumulation_steps == 0:
            optimizer.step()
            model.zero_grad()
            current = time.time()
            if (current - start) > TIME_IN_SECONDS:
                done = True
        if done:
            break

#saving model
torch.save(model.state_dict(), Path("./model/"))
```
